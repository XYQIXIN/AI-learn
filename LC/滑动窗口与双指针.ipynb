# 一：定长滑动窗口

窗口右端点在 i 时，由于窗口长度为 k，所以窗口左端点为 i−k+1。

我总结成三步：**入 - 更新 - 出**。

1. **入**：下标为 i 的元素进入窗口，更新相关统计量。如果窗口左端点 i−k+1<0，则尚未形成第一个窗口，重复第一步。
2. **更新**：更新答案。一般是更新最大值/最小值。
3. **出**：下标为 i−k+1 的元素离开窗口，更新相关统计量，为下一个循环做准备。

```python
# 题目1：给你字符串s和整数k，请返回字符串s中长度为k的单个子字符串中可能包含的最大元音字母数。

class Solution:
    def maxVowels(self, s: str, k: int) -> int:
        ans = vowel = 0  # ans: 最大元音数，vowel: 当前窗口的元音数
        for i, c in enumerate(s):  # i: 索引，c: 当前字符
            # 1. 进入窗口：如果当前字符是元音，计数器 +1
            if c in "aeiou":
                vowel += 1
            
            left = i - k + 1  # 计算窗口左端点
            # 窗口尚未形成（左端点 < 0），继续让元素进入窗口
            if left < 0:
                continue
            
            # 2. 更新答案：更新最大元音数
            ans = max(ans, vowel)
            # 剪枝优化：已经达到理论最大值 k，提前结束
            if ans == k:
                break
            
            # 3. 出窗口：左端点字符离开窗口，如果是元音，计数器 -1
            if s[left] in "aeiou":
                vowel -= 1
        
        return ans
```

题目2：给你一个由 n 个元素组成的整数数组 nums 和一个整数 k，请你找出平均数最大且 长度为 k 的连续子数组，并输出该最大平均数。

```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        max_s = -inf  # 初始化为负无穷，保证第一次比较时能更新（找最大值的标准做法）
        s = 0  # 当前窗口的和
        for i, c in enumerate(nums):  # i: 索引，c: 当前元素
            # 1. 进入窗口：累加当前元素
            s += c
            
            left = i - k + 1  # 计算窗口左端点
            # 窗口尚未形成（左端点 < 0），继续让元素进入窗口
            if left < 0:
                continue
            
            # 2. 更新答案：更新最大和
            max_s = max(max_s, s)
            
            # 3. 出窗口：左端点元素离开窗口，从和中减去
            s -= nums[left]  # nums[left] 即 nums[i-k+1]
        
        return max_s / k  # 返回最大平均数
```

题目3：给你一个下标从 0 开始的数组 nums，数组中有 n 个整数，另给你一个整数 k。半径为 k 的子数组平均值是指：nums 中一个以下标 i 为中心且半径为 k 的子数组中所有元素的平均值，即下标在 i - k 和 i + k 范围（含 i - k 和 i + k）内所有元素的平均值。如果在下标 i 前或后不足 k 个元素，那么半径为 k 的子数组平均值是 -1。构建并返回一个长度为 n 的数组 avgs，其中 avgs[i] 是以下标 i 为中心的子数组的半径为 k 的子数组平均值。

**思路**：本题相当于一个长为 2k+1 的滑动窗口。

**三步法**：
1. **入**：下标为 i 的元素进入窗口，窗口元素和 s 增加 nums[i]。如果 i<2k 则重复第一步。
2. **更新**：记录答案 avgs[i−k]=⌊ s/(2k+1)⌋。i−k 是因为 i 对应子数组右端点，答案记录在子数组正中间。
3. **出**：下标为 i−2k 的元素离开窗口，窗口元素和 s 减少 nums[i−2k]。

```python
class Solution:
    def getAverages(self, nums: List[int], k: int) -> List[int]:
        avgs = [-1] * len(nums)  # 占位数组：-1 表示该位置无法形成完整窗口
        s = 0  # 窗口元素和
        for i, x in enumerate(nums):  # i: 索引，x: 当前元素
            # 1. 进入窗口：累加当前元素
            s += x
            
            # 窗口大小不足 2k+1（需要 i >= 2k 才能形成完整窗口）
            if i < 2 * k:
                continue
            
            # 2. 更新答案：记录在子数组中间位置 i-k
            avgs[i - k] = s // (2 * k + 1)  # 整数除法，去掉小数部分
            
            # 3. 出窗口：减去最左边的元素 nums[i-2k]
            s -= nums[i - 2 * k]
        
        return avgs
```

题目4：给你一个整数数组 arr 和两个整数 k 和 threshold。请你返回长度为 k 且平均值大于等于 threshold 的子数组数目。

```python
class Solution:
    def numOfSubarrays(self, arr: List[int], k: int, threshold: int) -> int:
        s = 0  # 窗口元素和
        ans = 0  # 满足条件的子数组数目
        for i, x in enumerate(arr):  # i: 索引，x: 当前元素
            # 1. 进入窗口：累加当前元素
            s += x
            
            # 窗口尚未形成（左端点 i-k+1 < 0）
            if i < k - 1:
                continue
            
            # 2. 更新答案：检查平均值是否 >= threshold
            # 优化：避免浮点数，两边乘以 k，即 s >= k * threshold
            if s >= k * threshold:
                ans += 1
            
            # 3. 出窗口：减去左端点元素 arr[i-k+1]
            s -= arr[i - k + 1]
        
        return ans
```

题目5：给你一个长度为 n 下标从 0 开始的字符串 blocks ，blocks[i] 要么是 'W' 要么是 'B' ，表示第 i 块的颜色。字符 'W' 和 'B' 分别表示白色和黑色。给你一个整数 k ，表示想要 连续黑色块的数目。每一次操作中，你可以选择一个白色块将它涂成黑色块。请你返回至少出现一次连续k个黑色块的 最少操作次数。

```python
class Solution:
    def minimumRecolors(self, blocks: str, k: int) -> int:
        count = 0        # 当前窗口中白色块 W 的数量（需要涂黑的次数）
        result = k       # 最多需要 k 次（全白的情况）
        
        for i, x in enumerate(blocks):  # i: 索引，x: 当前字符
            # 1. 进入窗口：如果当前是白色，count +1
            if x == 'W':
                count += 1
            
            # 计算窗口左端点（必须在 if 外面，保证每次都计算）
            left = i - k + 1
            
            # 窗口尚未形成（i 还不到 k-1）
            if left < 0:
                continue
            
            # 2. 更新答案：当前窗口需要 count 次操作，取最小值
            result = min(result, count)
            
            # 3. 出窗口：左端点移出，如果是白色，count -1
            if blocks[left] == 'W':
                count -= 1
        
        return result
```
